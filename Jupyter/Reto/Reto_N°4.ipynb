```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import drive
drive.mount('/content/drive')
```

```python
ruta = "/content/drive/MyDrive/Jupyter/csv/Titanic-Dataset.csv"

df = pd.read_csv(ruta)

df
```

```python
def contar_sobrevivientes(df):
    return df[df["Survived"] == 1].shape[0]

def contar_muertos(data):
    return df[df["Survived"] == 0].shape[0]

print("Sobrevivientes:", contar_sobrevivientes(df))
print("Muertos:", contar_muertos(df))
```

```python
sobrevivientes = contar_sobrevivientes(df)
muertos = contar_muertos(df)


datos = pd.Series({
    "Sobrevivientes": sobrevivientes,
    "Muertos": muertos})

datos.plot(kind="bar", xlabel="Categoría", ylabel="Cantidad", title="Supervivencia en el Titanic", color=["green", "red"])
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
