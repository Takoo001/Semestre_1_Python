```python
import numpy as np
import matplotlib.pyplot as plt
```

###Generando un Vector aleatorio de 12 numeros enteros (Desde 1 al 10) con Random

```python
vector = np.random.randint(1, 11, 12)
vector
```

####Obtencion de la Dimension de un Arreglo

```python
np.shape(vector)
```

####Tranformar el vector de 12 elementos en una matriz

```python
matriz = vector.reshape(4, 3)
matriz
```

####Volviendo a consultar la dimension pero de la Matriz

```python
np.shape(matriz)
```

##Conociendo matplotlib

####Creando un Grafico Simple

```python
plt.plot(vector, label = "Vector")
plt.title("Grafico del Vector")
plt.xlabel("Indice (Posicion)") # TITULO EJE X
plt.ylabel("Valores") # TITULO EJE Y

plt.grid(True) #PARA AGREGAR GRILLA AL GRAFICO (CUADRICULA)
plt.legend() #MOSTRAR UNA ETIQUETA QUE SE OBTIENE DE LA PROPIEDA LABEL DEL plt.plot


plt.show() #PERMITE VIZUALIZAR EL GRAFICO PERO NO ES NECESARIO PONERLO
```

####Crear dos vectores aleatorios en un mismo grafico


1.   Elemento de lista
2.   Elemento de lista



```python
vector1 = np.random.randint(1,11,12)
vector2 = np.random.randint(2,21,12)
```

```python
plt.plot(vector1, marker = "o", label = "Vector 1")
plt.plot(vector2, marker = "o", label = "Vector 2")
plt.title("Grafico del Vector")
plt.xlabel("Indice (Posicion)")
plt.ylabel("Valores")

plt.grid(True)
plt.legend()


plt.show()
```

###Creando un grafico de barra

```python
plt.bar(range(len(vector2)), vector1) #EL METODO BAR PERMITE PLOTEAR UN GRAFICO DE BARRAS
```
