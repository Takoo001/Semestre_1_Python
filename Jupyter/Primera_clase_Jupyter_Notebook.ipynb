```python
# Importando bibliotecas
import numpy as np
import pandas as pd
import random as rd
import sys
```

#####**Trabajando con lista**

```python
# Creando una lista de 10 elementos
lista = list(range(1,11))
lista
```

```python
# Transformando una lista a array (arreglo de numpy)
arreglo = np.array(lista)
arreglo
```

```python
# Promedio
promedio = np.mean(lista)
promedio
```

```python
# Consultando la cantidad de memoria ocupada por estructura
memoria_lista = sys.getsizeof(lista)
memoria_arreglo = arreglo.nbytes

print(f"Memoria ocupada por la lista: {memoria_lista} Bytes")
print(f"Memoria ocupada por el array: {memoria_arreglo} Bytes")
```

####**Aplicando el paquete random**


```python
np.random.randint(1,100) # Generando numeros aleatorios del 1 al 100
```

```python
np.random.rand() # Genera numeros aleatorios del rango 0 a 1
```

```python
np.random.randint(10) # Genera numeros aleatorios del rango de 0 a 1 con 10 elementos pero mas parecido a una matriz
```

####**Otros metodos de numpy**

```python
print("Shape:", arreglo.shape) # Dimensiones
print("Dimensiones:", arreglo.ndim) # Numero de ejes
print("Tama√±o:", arreglo.size) # Total de elementos
print("Tipo de dato:", arreglo.dtype) # Tipo almacenado
print("Bytes/elem:", arreglo.itemsize) # Tama√±o en bytes de cada elemento
```

####**Transformando el arreglo de Int a Float**

```python
new_arreglo= arreglo.astype(np.float32) # Especificando que sea de 32 bits (ocupa menos memoria)
new_arreglo
```

El metodo astype() transforma los elementos del arrelgo a un tipo determinado de dato numerico
